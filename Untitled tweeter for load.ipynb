```python
# this block is used to stream timed data, specific to a topic
'''
Twitter Straming API download script


from __future__ import absolute_import

from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream
import sys
import time
import json

tlimit = 10     # set the time limit (in seconds) for how long to run the streamer

keyword_list = ['xxx'] # Search terms

#~~~~~~~~~~~~~~~~~~

start_time = time.time() # Set the current system time as "start_time" so we can calculate how much time has passed later

class listener(StreamListener):
    """ A listener handles tweets are the received from the stream.
    This is a basic listener that just prints received tweets to stdout.
    """

    def __init__(self, start_time, time_limit=30): # default values; passing new time_limit later overwrites this time_limit
        self.time = start_time # Push start_time to variable timem for simpler coding terms
        self.limit = time_limit # Push time_limit to limit for same reason
    
    def on_data(self, data):
        
        while (time.time() - self.time) < self.limit: # Basic arithmetic to ensure that the time passed is less than the limit we desire (current time - start_time should be less than time_limit)
            
            try:
                json_load = json.loads(data)
                texts = json_load['text']
                coded = texts.encode('utf-8')
                s = str(coded)
                print(s[2:-1])
                
                return True # Without this line, the script will simply repeat the first tweet it finds ad infinitum. Don't forget this line.
                
            except BaseException as e: # If an exception occurs, capture the code. 
                print('failed ondata,', str(e)) # Print "failed ondata" and the error code
                time.sleep(5) # Wait 5 seconds before doing anything else
                pass # Move on
        
        sys.exit() # Exit the function after the while loop ends
        
        #print(data)
        #return True

    def on_error(self, status):
        print(status) # If all goes well, we won't see this print


auth = OAuthHandler(consumer_key, consumer_secret) #OAuth object
auth.set_access_token(access_token, access_token_secret) 
 
twitterStream = Stream(auth, listener(start_time, time_limit=tlimit)) #initialize Stream object with a time out limit
twitterStream.filter(track=keyword_list, languages=['en'])  #call the filter method to run the Stream Object
'''
```

```python
DISPLAY_EVERY = 20

stop = set(stopwords.words('english'))  # predefined list of "uninteresting" words

counter = Counter()

def nlargest(n, word_scores):
    """ Wrapper around heapq to return the n words with the largest count."""
    # word_scores: index 0=>word, 1=>count
    return heapq.nlargest(n, word_scores, key=lambda x: x[1])

try:
    # islice allows you to get some # of values out of a generator.
    for k, tweet in enumerate(islice(tweet_generator(), 1000)):
        for word in tweet.lower().encode('utf-8').split():  # lowercase, split by whitespace
        #for word in re.split('\W+', tweet.lower()): #if you want to manually exclude some general words, use regular expresseion here
            if word not in stop:  # avoid uninteresting words
                counter[word] += 1
        if k % DISPLAY_EVERY == (DISPLAY_EVERY - 1):  # only update text every so often
            # \r will overwrite updates, rather than listing them out
            # one on each newline
            sys.stdout.write("\r" + str(nlargest(20, counter.items())))
except KeyboardInterrupt:
    pass  # allow ctrl-c to exit the loop
finally:

    df = pd.DataFrame(nlargest(20, counter.items()), columns=['words', 'count'])
    df.set_index('words').plot(kind='bar')
```
